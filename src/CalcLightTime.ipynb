## GET Time

Calculates the delta time between to date and time objects.

```python
import datetime
import pandas as pd
```

```python
df = pd.read_csv('LightStatus.csv') # reads the csv file into a Pandas Data Frame.
```

```python
dt = df['Date'] # Get a copy of the Date/Time column as a list of strings
```

```python
format = '%y/%m/%dT%H:%M:%S]' # The date format used to convert to a date/time object
```

The next 2 lines of code convert the date/time string values to days hours and seconds since since December 30, 1899. Math operations can be performed on the objects.

```python
dtstart = datetime.datetime.strptime(dt[1], format)
dtstart
```

```python
dtend = datetime.datetime.strptime(dt[20], format)
dtend
```

```python
dtdelta = dtend - dtstart
```
